```python
import sys
import pygame
import random
import time

# 게임 화면 크기
WINDOW_WIDTH = 1000
WINDOW_HEIGHT = 550
pygame.init()

# 색상
GRAY = (150, 150, 150)
BLACK = (0, 0, 0)

# 소스 디렉토리

DIRBEATS = 'C:/Users/sunsum/Documents/opensource/teamproject/beat/'

# 기본 변수
SCORE = 0
BEAT_COUNT = 78

# bground
bground_img = pygame.image.load('bground.png')
bground = pygame.transform.scale(bground_img, (1000, 550))

# music
music = pygame.mixer.Sound('backmusic.wav')  # 배경음악
do = pygame.mixer.Sound('do.wav')  # 핸드벨
re = pygame.mixer.Sound('re.wav')
mi = pygame.mixer.Sound('mi.wav')
pa = pygame.mixer.Sound('pa.wav')
sol = pygame.mixer.Sound('sol.wav')
la = pygame.mixer.Sound('la.wav')
si = pygame.mixer.Sound('si.wav')
ddo = pygame.mixer.Sound('ddo.wav')

time.sleep(3)  # 시간지연 3초
music.play()  # 배경음악 1회실행

# beat오브젝트를 저장할 리스트
BEATS = []



class Beat:
    beat_image = ['1.png', '2.png', '3.png', '4.png', '5.png', '6.png', '7.png', '8.png']

    def __init__(self, x=0, y=0, dx=0, dy=0):
        self.image = ''
        self.x = x
        self.y = y
        self.dx = dx
        self.dy = dy
        self.rect = ''

    def load_beat(self, p=""):
        if p == "p":
            # 플레이어
            self.image = pygame.image.load(DIRBEATS + "player.png")
            self.image = pygame.transform.scale(self.image, (100, 100))
            self.rect = self.image.get_rect()
            self.rect.x = self.x
            self.rect.y = self.y

        else:
            # 비트 40개의 이미지중에서 랜덤으로 선택한다. 
            self.image = pygame.image.load(DIRBEATS + random.choice(self.beat_image))
            self.rect = self.image.get_rect()

            carwidth = 100
            carheight = 100

            self.image = pygame.transform.scale(self.image, (carwidth, carheight))
            self.rect.width = carwidth
            self.rect.height = carheight

            # 생성 위치 
            self.rect.x = self.x
            self.rect.y = self.y



    def draw_beat(self):
        SCREEN.blit(self.image, [self.rect.x, self.rect.y])

    # x 좌표 이동 - 플레이어의 움직임

    def move_x(self):
        self.rect.x = self.dx

    # y 좌표 이동 - 비트 y 움직임

    def move_y(self):
        self.rect.y += self.dy

    # 비트 충돌 감지
    # distance : 오른쪽, 왼쪽, 아래쪽, 위쪽 이미지의 간격 설정

    def check_collision(self, car, distance=0):
        if (self.rect.top + distance < car.rect.bottom) and (car.rect.top < self.rect.bottom - distance) and (
                self.rect.left + distance < car.rect.right) and (car.rect.left < self.rect.right - distance):
            return True
        else:
            return False


def main():
    global SCREEN, BEAT_COUNT, WINDOW_WIDTH, WINDOW_HEIGHT, SCORE

    # pygame 초기화 및 스크린 생성

    pygame.init()
    SCREEN = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT),
                                     pygame.HWSURFACE | pygame.DOUBLEBUF | pygame.RESIZABLE)
    pygame.display.set_caption("DO NOT CRY HAND BELL")
    # 투명 아이콘은 32 x 32 로 해야 적용 됨
    windowicon = pygame.image.load(DIRBEATS + '1.png').convert_alpha()
    pygame.display.set_icon(windowicon)

    # bellcatimage
    bell1_img = pygame.image.load('belldown.png')
    bell2_img = pygame.image.load('bellup.png')

    # 14초후 비트 떨어지기
    time.sleep(14)
    clock = pygame.time.Clock()

    # player class 지정
    player = Beat(50, 450, 0, 0)
    player.load_beat("p")

    # beat 좌표 class에 추가
    beat = Beat(300, 0, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(400, -100, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -200, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -300, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 미파솔솔
    beat = Beat(600, -500, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(700, -600, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(800, -700, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(800, -800, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 라시도도
    beat = Beat(300, -1000, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(400, -1100, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -1200, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -1300, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 미파솔솔
    beat = Beat(600, -1450, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -1550, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(400, -1650, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(400, -1750, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 라솔파파
    beat = Beat(300, -1950, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -2050, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(100, -2150, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(300, -2250, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 미솔도미
    beat = Beat(200, -2400, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(400, -2500, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(200, -2700, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(100, -2800, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 레파레도-----------------------
    beat = Beat(300, -3800, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(400, -3900, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -4000, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -4100, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 미파솔솔
    beat = Beat(600, -4300, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(700, -4400, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(800, -4500, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(800, -4600, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 라시도도
    beat = Beat(300, -4800, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(400, -4900, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -5000, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -5100, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 미파솔솔
    beat = Beat(600, -5300, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -5400, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(400, -5500, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(400, -5600, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 라솔파파-
    beat = Beat(300, -5750, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -5850, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(100, -5950, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(300, -6050, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 미솔도미
    beat = Beat(200, -6250, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(400, -6350, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(200, -6550, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(100, -6650, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 레파레도------------------------
    beat = Beat(300, -7500, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(400, -7590, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -7700, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -7800, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 미파솔솔
    beat = Beat(600, -8000, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(700, -8100, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(800, -8200, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(800, -8300, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 라시도도
    beat = Beat(300, -8500, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(400, -8600, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -8700, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -8800, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 미파솔솔
    beat = Beat(600, -8950, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -9050, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(400, -9150, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(400, -9250, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 라솔파파
    beat = Beat(300, -9450, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(500, -9550, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(100, -9650, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(300, -9750, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 미솔도미
    beat = Beat(200, -9900, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(400, -10000, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(700, -10100, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(800, -10200, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 레파시도---
    beat = Beat(800, -10300, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(800, -10400, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(800, -10500, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(800, -10600, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 도도도도
    beat = Beat(500, -10800, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    beat = Beat(100, -10900, 0, 10)
    beat.load_beat()
    BEATS.append(beat)
    # 솔도
    playing = True

    while playing:
        SCREEN.fill(GRAY)
        SCREEN.blit(bground, (0, 0))
        SCREEN.blit(bell1_img, (100, 100))
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                playing = False
                pygame.quit()
                sys.exit()

            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_a:
                    do.play()
                    player.dx = 100
                    SCREEN.blit(bell2_img, (100, 100))
                elif event.key == pygame.K_s:
                    re.play()
                    player.dx = 200
                elif event.key == pygame.K_d:
                    mi.play()
                    player.dx = 300
                elif event.key == pygame.K_f:
                    pa.play()
                    player.dx = 400
                elif event.key == pygame.K_g:
                    sol.play()
                    player.dx = 500
                elif event.key == pygame.K_h:
                    la.play()
                    player.dx = 600
                elif event.key == pygame.K_j:
                    si.play()
                    player.dx = 700
                elif event.key == pygame.K_k:
                    ddo.play()
                    player.dx = 800

        player.load_beat("p")
        player.move_x()
        player.draw_beat()
        for i in range(BEAT_COUNT):
            BEATS[i].draw_beat()
            BEATS[i].rect.y += BEATS[i].dy
            if i == BEAT_COUNT:
                break

        for i in range(BEAT_COUNT):
            if player.check_collision(BEATS[i], 0):
                # 부딪쳤을 경우 빠르게 내려가기
                if player.rect.x == BEATS[i].rect.x:
                    BEATS[i].rect.y += 30
                    SCORE += 10

        print(SCORE)
        if SCORE >= 3800:
            SCREEN.blit(bground, (0, 0))

        #글씨
        font_01 = pygame.font.SysFont("FixedSsy", 30, True, False)
        text_score = font_01.render("Score : " + str(SCORE), True, BLACK)
        SCREEN.blit(text_score, [15, 15])
        pygame.display.flip()
        
        # 초당 프레임 설정
        clock.tick(25)


if __name__ == '__main__':
    main()
```
